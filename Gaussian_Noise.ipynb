# **Generate synthetic values**

Gaussian Noise (GN): this algorithm adds Gaussian noise to the interpolation and undersamples abundant values. It ensures that the distribution of the synthetic data does not differ extremely from the original minority sample. The code is borrowed from https://github.com/paobranco/ImbalancedLearningRegression.git 

```python
## conduct GN
data_train_gn = iblr.gn(
    
    data = data_train_ig,  ## pandas dataframe
    y = 'seq_TEMP',  ## string ('header name'),
    rel_thres = 0.7, ## relevance threshold considered rare
    under_samp = False
)
```

# **t-SNE for selected features**

T-SNE is applied to visualize the distribution of the resampled testing data into two-dimensional scale.

```python
scaler = StandardScaler()

data_train_gn_sc = scaler.fit_transform(data_train_gn.drop(['seq_TEMP'], axis=1))

embedding = TSNE(n_components=2)

# transform on features
new_tsne_gn = embedding.fit_transform(data_train_gn_sc)

# reconstruct dataframe
new_tsne_gn = pd.DataFrame(new_tsne_gn, columns=[f"TSNE{i+1}" for i in range(new_tsne_gn.shape[1])])
new_tsne_gn["seq_TEMP"] = data_train_gn["seq_TEMP"]
```

```python
#Code for displaying principal components
#Code for displaying principal components
sns.set_style("white")
plt.scatter(new_tsne_gn.TSNE1, new_tsne_gn.TSNE2, s=5)
plt.xlabel("t-SNE 1")
plt.ylabel("t-SNE 2")
plt.title("Gaussian Noise")
plt.show()
```
