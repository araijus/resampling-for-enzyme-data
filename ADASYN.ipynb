# **Generate synthetic values**

Adaptive Synthetic Sampling and the Stacking Algorithm (ADASYN): similar to SMOTE it generates synthetic data based on 5-nearest neighbors but based on data density. It considers deviations of synthesized samples from original samples and adds more data in the areas with low density of rare values. The code is borrowed from https://github.com/paobranco/ImbalancedLearningRegression.git 

```python
## conduct ADASYN
data_train_adasyn = iblr.adasyn(
    
    data = data_train_ig,  ## pandas dataframe
    y = 'seq_TEMP'  ## string ('header name'),
)
```
